```python
import cv2
import matplotlib.pyplot as plt
import cvlib as cv
from cvlib.object_detection import draw_bbox
import globe
```

```python
velo = cv2.imread('velo.jpg')
bbox, label, conf = cv.detect_common_objects(velo)
output_image = draw_bbox(velo, bbox, label, conf)
plt.imshow(output_image)
plt.show()
```

```python
#charger des images en couleurs sous forme d'images en niveaux de gris ou sous forme d'images en couleur
#Pour les images chargées en couleurs BGR, les convertir en HSV et en YUV

velo = cv2.imread('velo.jpg')
# velo = cv2.cvtColor(velo, cv2.COLOR_BGR2GRAY)
# velo1 = cv2.cvtColor(velo, cv2.COLOR_BGR2HSV)
velo3 = cv2.cvtColor(velo, cv2.COLOR_BGR2YUV)
plt.imshow(velo3)
print('Original Dimensions : ',velo.shape)
```

```python
scale_percent = 60 # percent of original size
width = int(velo.shape[1] * scale_percent / 100)
height = int(velo.shape[0] * scale_percent / 100)
dim = (width, height)
```

```python
# resize image
resized = cv2.resize(velo, dim, interpolation = cv2.INTER_AREA)
```

```python
print('Resized Dimensions : ',resized.shape)
 
cv2.imshow("Resized image", resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
